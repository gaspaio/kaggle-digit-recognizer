```python
%matplotlib inline

from matplotlib import pyplot as plt

import seaborn as sns  
import datetime as dt
import numpy as np
import pandas as pd

from sklearn.decomposition import TruncatedSVD
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.utils.random import sample_without_replacement

from sklearn.metrics import confusion_matrix, accuracy_score

sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (10, 6)})
```

# Data

```python
train = pd.DataFrame.from_csv('data/train.csv', index_col=None)
test = pd.DataFrame.from_csv('data/test.csv', index_col=None)

Y = train['label'].values
X = train.drop('label', axis=1).values
X_test = test.values

print "Training data is composed of {} {} pixel images".format(X.shape[0], X.shape[1])
print "Test data is composed of {} {} pixel images".format(X_test.shape[0], X_test.shape[1])
```

```python
# The training data is big. We'll subsample it for our tests.
idxs = sample_without_replacement(X.shape[0], 1000)
X = X[idxs].copy()
Y = Y[idxs].copy()

# Is it balanced ? (should have 10% each)
print np.bincount(Y)*1.0/Y.shape[0]
```

# KNN

### Basic KNN with arbitrary params 

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.3)

clf = KNeighborsClassifier(n_neighbors=5).fit(X_train,Y_train)

# In KNN, fit doesn't do much, just stores the labeled data. When we predict, the training data is used to vote for an output label.
Y_val_hat = clf.predict(X_val)
```

```python
cm = confusion_matrix(Y_val, Y_val_hat)
print cm
print "Accuracy with validation set and k=5: {:.2f}".format(accuracy_score(Y_val, Y_val_hat))
```
